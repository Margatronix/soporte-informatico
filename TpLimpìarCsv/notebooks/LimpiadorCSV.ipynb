```python
import pandas as pd
import numpy as np

def arreglar_columnas(valor):
    for i in range(len(valor)):
        valor[i] = valor[i].replace(" ","")
    return valor

def resolver_espacios(valor): #Funcion que elimina el espacio de delante y si el valor es nan o vacio, le pone None para que el dropna lo saque
    valor = str(valor)
    if valor[0] == ' ':
        valor = valor[1:]
    if valor == '':
        valor = np.nan
    if valor == 'nan':
        valor = np.nan
    return valor

def resolver_altura(valor):
    valor = str(valor)
    for caracter in valor:
        if not caracter.isnumeric() and caracter != ".":
            valor = valor.replace(caracter,'')
    return valor

def resolver_peso(valor):
    valor = str(valor)
    for caracter in valor:
        if not caracter.isnumeric() and caracter != ".":
            valor = valor.replace(caracter,'')
    valor = float(valor) #Para pasar de str a int, tenes que pasar por float antes
    valor = int(valor)
    return valor

def resolver_promedio(valor):
    valor = float(valor)
    return valor

def resolver_sexo(valor):
    valor = str(valor).lower()
    valor = valor[0]
    if valor=='h' or valor=='m':
        return valor
    elif valor=='n':
        return 'nb'
    else:
        return np.nan

def resolver_LU(valor):
    valor = str(valor)
    i = 0
    for caracter in valor:
        if not caracter.isnumeric():
            if i==0:
                valor = valor.replace(caracter,'/')
                i=1
            else:
                valor = valor.replace('//','/')
    return valor

def resolver_observacion(valor):
    valor = str(valor)
    valor =valor.lower()
    if valor != "ninguna":
        if valor == "hipertenso" or valor == "hiper tension":
            valor = "hipertension"
        if valor == "dolor cabeza":
            valor = "dolor de cabeza"
        elif valor != "hipertension":
            valor = "ninguna"
    return valor

csv = pd.read_csv("../data/datos.csv")

csvColumnas = list(csv.columns)
csv.columns = arreglar_columnas(csvColumnas)

csv = csv.applymap(resolver_espacios)
csv.dropna(inplace=True)

csv['Altura'] = csv['Altura'].apply(resolver_altura)

csv['Peso'] = csv['Peso'].apply(resolver_peso)

csv['Promedio'] = csv['Promedio'].apply(resolver_promedio)

csv['Sexo'] = csv['Sexo'].apply(resolver_sexo)

csv['LU'] = csv['LU'].apply(resolver_LU)

csv['Observacion'] = csv['Observacion'].apply(resolver_observacion)

csv.dropna(inplace=True)
print(csv)


def resolver_viajeros(valor):
    valor = str(valor)
    if valor == '':
        valor = "null"
    return valor

def resolver_observaciones(valor):
    valor = str(valor)
    if valor == '""':
        valor = "ninguna"
    return valor
```
